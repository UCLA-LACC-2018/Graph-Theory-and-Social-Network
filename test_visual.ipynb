```python
#setting parameters
import random
import math
p=0.05#the threshold of probability 
N=100#number of nodes in the graph
disthres=0.8
#replace LACC2018 with LACC2018_E while testing your own code
from LACC2018 import matrix_load,print_degrees,shortest_path,detect_cycle
```

```python
#generating the tested graph
#the existence of the edge between any two nodes is subjected to an uniform distribution on (0,1)
a=[0]*N
b=[0]*N
for x in range(N):
    a[x]=random.uniform(0,1)
    b[x]=random.uniform(0,1)
mat=[]#adjacent matrix
for x in range(N):
	mat.append([0]*N)
for x in range(N):
	for y in range(x+1,N):
		if random.uniform(0,1)<p and math.sqrt((a[x]-a[y])**2+(b[x]-b[y])**2)<disthres:
			mat[x][y]=1
			mat[y][x]=1
```

```python
import matplotlib.pyplot as plt
plt.scatter(a,b,label="nodes",color="green",marker="*",s=30)
for x in range(N):
    for y in range(x+1,N):
        if mat[x][y]==1:
            plt.plot([a[x],a[y]],[b[x],b[y]],color="grey")

#exercise2: find the shortest path from arbitrarily two nodes
node1=2;
node2=8;
plt.scatter(a[node1],b[node1],label="nodes",color="blue",marker="o",s=30)
plt.scatter(a[node2],b[node2],label="nodes",color="blue",marker="o",s=30)
path=shortest_path(mat,node1,node2)
if path!=[]:
    for t in range(len(path)-1):
        x=path[t]
        y=path[t+1]
        if mat[x][y]==1:
            plt.plot([a[x],a[y]],[b[x],b[y]],color="red")
```

```python
#exercise3: detect an output the cycle in a random graph
plt.scatter(a,b,label="nodes",color="green",marker="*",s=30)
for x in range(N):
    for y in range(x+1,N):
        if mat[x][y]==1:
            plt.plot([a[x],a[y]],[b[x],b[y]],color="grey")
cycle=detect_cycle(mat)
if cycle!=[]:
    for t in range(len(cycle)-1):
        x=cycle[t]
        y=cycle[t+1]
        if mat[x][y]==1:
            plt.plot([a[x],a[y]],[b[x],b[y]],color="blue")    
    x=cycle[len(cycle)-1]
    y=cycle[0]
    if mat[x][y]==1:
        plt.plot([a[x],a[y]],[b[x],b[y]],color="blue")
```
